```python
# Ganti nilai inf pada trailing PE dengan NaN
stock_fundamental['trailing_pe'].replace([np.inf, -np.inf], np.nan, inplace=True)

# EPS terlalu besar atau kecil secara ekstrem
stock_fundamental['trailing_eps'] = stock_fundamental['trailing_eps'].mask(
    stock_fundamental['trailing_eps'].abs() > 1000
)

# PEG ratio tidak wajar (< -10 atau > 100)
stock_fundamental['peg_ratio'] = stock_fundamental['peg_ratio'].mask(
    (stock_fundamental['peg_ratio'] > 100) | (stock_fundamental['peg_ratio'] < -10)
)

# Debt to equity ratio ekstrem (> 1000)
stock_fundamental['debt_to_equity'] = stock_fundamental['debt_to_equity'].mask(
    stock_fundamental['debt_to_equity'] > 1000
)

# Growth metrics ekstrem (di luar -100% sampai 1000%)
growth_cols = ['earnings_growth', 'revenue_growth', 'earnings_quarterly_growth']
for col in growth_cols:
    stock_fundamental[col] = stock_fundamental[col].mask(
        (stock_fundamental[col] > 1000) | (stock_fundamental[col] < -100)
    )
stock_fundamental.describe()
```

## ðŸš§ Penanganan Outliers pada Data Fundamental Saham

Setelah melakukan eksplorasi awal terhadap statistik deskriptif, ditemukan bahwa beberapa kolom pada data memiliki nilai-nilai ekstrem (outliers) yang **secara logika bisnis tidak masuk akal atau sangat jarang terjadi**. Outliers semacam ini dapat mengganggu proses analisis dan menghasilkan insight yang tidak akurat, terutama jika digunakan dalam visualisasi atau pemodelan.

Berikut ini adalah beberapa penyesuaian yang dilakukan:

### ðŸ“Œ 1. Kolom `trailing_pe`
- Nilai `inf` atau `-inf` sering muncul karena pembagian dengan EPS nol atau sangat kecil.
- **Solusi:** Nilai `inf` dan `-inf` diganti dengan `NaN` agar dapat ditangani pada proses selanjutnya.

### ðŸ“Œ 2. Kolom `trailing_eps`
- Ditemukan nilai EPS yang sangat ekstrem (di atas 1000 atau di bawah -1000), yang kemungkinan merupakan kesalahan input atau kondisi luar biasa.
- **Solusi:** EPS dengan nilai absolut lebih dari 1000 diubah menjadi `NaN`.

### ðŸ“Œ 3. Kolom `peg_ratio`
- PEG ratio secara umum wajar di antara nilai positif kecil hingga sedang. Nilai di bawah -10 atau di atas 100 dianggap tidak wajar.
- **Solusi:** Diberikan batas bawah dan atas untuk menghapus nilai ekstrem tersebut.

### ðŸ“Œ 4. Kolom `debt_to_equity`
- Rasio ini bisa sangat tinggi untuk sektor tertentu, tetapi nilai di atas 1000 dianggap tidak wajar secara umum.
- **Solusi:** Nilai di atas 1000 diubah menjadi `NaN`.

### ðŸ“Œ 5. Kolom `growth` (earnings & revenue growth)
- Pertumbuhan lebih dari 1000% atau penurunan di bawah -100% dianggap terlalu ekstrem dan jarang ditemukan dalam praktik nyata.
- **Solusi:** Nilai-nilai tersebut disaring menjadi `NaN` karena dianggap outliers.

### âœ… Tujuan
Langkah ini bertujuan untuk meningkatkan kualitas data dan mencegah nilai-nilai ekstrem tersebut mendominasi hasil analisis. Outliers yang tidak relevan atau tidak realistis dapat menyebabkan distorsi pada metrik agregat seperti rata-rata, korelasi, atau mempengaruhi performa model jika digunakan untuk pembelajaran mesin.

## ðŸ” Penanganan Missing Values

Pada tahap pembersihan data (data cleaning), ditemukan cukup banyak nilai kosong (missing values) pada beberapa kolom fundamental seperti:

- `trailing_pe`
- `trailing_eps`
- `peg_ratio`
- `ebitda_billionUSD`
- `debt_to_equity`
- `earnings_growth`
- `revenue_growth`, dll.

Alih-alih melakukan imputasi (pengisian nilai kosong), diputuskan untuk **menghapus baris-baris yang memiliki nilai kosong** pada kolom-kolom penting tersebut. Keputusan ini didasarkan pada beberapa pertimbangan:

### ðŸ§  Alasan Menghapus Missing Values

1. **Berkaitan dengan data keuangan yang sensitif**
   > Nilai seperti `EPS`, `EBITDA`, atau `PEG Ratio` bersifat spesifik untuk masing-masing perusahaan. Mengisinya dengan rata-rata atau median bisa merusak makna asli data dan menyebabkan bias dalam analisis.

2. **Missing values kemungkinan besar tidak terjadi secara acak**
   > Banyak nilai yang hilang karena perusahaan belum mencetak laba, baru IPO, atau tidak relevan untuk sektor tertentu. Maka, imputasi justru bisa menyamarkan perbedaan penting antar perusahaan.

3. **Ukuran data tetap mencukupi**
   > Setelah pembersihan, jumlah data masih cukup besar untuk dianalisis tanpa kehilangan representasi.

### âœ… Kesimpulan

Maka, strategi yang digunakan adalah **drop baris dengan missing values** pada kolom-kolom fundamental utama agar kualitas data tetap terjaga dan analisis yang dilakukan bersifat valid serta dapat dipertanggungjawabkan.

## ðŸ” Membentuk Variabel Baru dari Rasio Fundamental Saham

Dalam analisis fundamental saham, data mentah seperti pendapatan, laba, dan utang seringkali belum cukup informatif. Oleh karena itu, diperlukan proses **rekayasa fitur (feature engineering)** untuk membentuk **rasio-rasio keuangan baru** yang dapat memberikan wawasan lebih dalam mengenai kesehatan finansial dan valuasi perusahaan.

Pada tahap ini, sejumlah variabel baru dihitung berdasarkan kombinasi dari kolom-kolom yang telah tersedia. Berikut adalah penjelasan setiap indikator yang dibentuk:

---

### ðŸ“Œ 1. Price to Sales Ratio (P/S Ratio)

**Definisi:**  
Rasio ini menunjukkan seberapa besar harga saham investor bersedia bayar untuk setiap satu dolar pendapatan yang dihasilkan perusahaan. Cocok untuk menilai perusahaan yang belum menghasilkan laba, namun memiliki potensi pertumbuhan pendapatan yang besar.

**Interpretasi:**  
- Nilai P/S tinggi â†’ Saham dianggap mahal terhadap pendapatannya.  
- Nilai P/S rendah â†’ Saham undervalued atau efisiensi pendapatan tinggi.


---

### ðŸ“Œ 2. Net Debt to EBITDA

**Definisi:**  
Mengukur kemampuan perusahaan dalam melunasi utangnya menggunakan pendapatan operasional sebelum bunga, pajak, depresiasi, dan amortisasi (EBITDA). Ini mencerminkan tingkat leverage perusahaan secara lebih realistik.

**Interpretasi:**  
- Nilai < 3 â†’ Umumnya dianggap sehat.  
- Nilai tinggi â†’ Perusahaan berisiko terhadap tekanan utang.

---

### ðŸ“Œ 3. Revenue to Debt Ratio

**Definisi:**  
Rasio ini menunjukkan berapa kali total pendapatan perusahaan dapat menutupi total utangnya. Sangat berguna untuk melihat kekuatan pendapatan terhadap beban kewajiban.

**Interpretasi:**  
- Nilai > 1 â†’ Pendapatan lebih tinggi dari utang (positif).  
- Nilai < 1 â†’ Utang melebihi pendapatan (berisiko).


---

### ðŸ“Œ 4. Earnings Yield (%)

**Definisi:**  
Menggambarkan return yang dihasilkan dari laba bersih terhadap harga saham. Ini adalah kebalikan dari Price to Earnings Ratio dan memungkinkan perbandingan dengan instrumen seperti obligasi.

**Interpretasi:**  
- Earnings Yield tinggi â†’ Saham cenderung undervalued.  
- Earnings Yield rendah â†’ Saham bisa jadi overvalued atau pertumbuhannya tinggi.


---

### ðŸ“Œ 5. EBITDA Margin (%)

**Definisi:**  
Menunjukkan seberapa besar laba operasional (EBITDA) yang dihasilkan dari setiap satu dolar pendapatan. Metrik ini menggambarkan efisiensi operasional perusahaan.

**Interpretasi:**  
- EBITDA Margin tinggi â†’ Biaya operasional efisien, daya saing tinggi.  
- EBITDA Margin rendah â†’ Perusahaan kurang efisien atau biaya tinggi.


---

### ðŸ“Œ Kesimpulan

Pembuatan variabel-variabel baru ini bertujuan untuk membantu proses analisis data saham menjadi lebih terstruktur dan informatif. Dengan rasio-rasio tambahan ini, analis dapat menilai kondisi dan valuasi perusahaan dengan pendekatan yang lebih menyeluruh dan relevan terhadap kondisi pasar.
